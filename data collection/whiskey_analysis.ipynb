```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


data= pd.read_csv("bottlers.csv")
ax=plt.figure(figsize=(5,10))
sns.scatterplot(y=data['Country'],x=data['Rating'])
data=data.drop('Unnamed: 7', axis=1)
```

```python

data=data[data['Rating'].notna()]
data=data[data['Rating']!=0]
sns.violinplot(x=data['Rating'])
```

```python
ax=plt.figure(figsize=(10,10))
my_order = data.groupby(by=["Country"])["Rating"].median().sort_values().iloc[::-1].index
sns.violinplot(y=data['Country'],x=data['Rating'],order=my_order)
```

```python
ax=plt.figure(figsize=(10,10))
my_order = data.groupby(by=["Country"])["Whiskies"].sum().sort_values().iloc[::-1].index
sns.barplot(y=data['Country'],x=data['Whiskies'],order=my_order)
```

```python
my_order2 = data.groupby(by=["Country"])["Whiskies"].sum().sort_values().iloc[::-1]
ax=plt.figure(figsize=(10,10))

sns.barplot(y=my_order2.index,x=my_order2)
```

```python
my_order2 = data.groupby(by=["Country"])["Rating"].mean().sort_values().iloc[::-1]
ax=plt.figure(figsize=(10,10))

sns.barplot(y=my_order2.index,x=my_order2)
```
